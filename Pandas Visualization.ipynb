# Visualization in Pandas

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as  plt
```

```python
df = pd.read_csv(r"D:\Python Tuts\Pandas Tutorial\Ice Cream Ratings.csv")
df = df.set_index('Date')
df
```

```python
df.plot(kind = 'line') #default
df.plot(kind = 'line', subplots = True)
```

```python
df.plot(kind = 'line', title = 'Ice Cream Ratings', xlabel = 'Daily Ratings', ylabel = 'Scores')
```

```python
#df.plot(kind = 'bar', stacked = True)
#df['Flavor Rating'].plot(kind = 'bar', stacked = True)
df.plot.barh(stacked = True)
```

```python
#df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating')
#df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating', s = 200)
df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating', s = 200, c = 'Yellow')
```

```python
#df.plot.hist()
df.plot.hist(bins = 20) #bins are distributions
```

```python
df.boxplot()
```

```python
#df.plot.area()
df.plot.area(figsize = (10,5))
```

```python
#df.plot.pie(y='Flavor Rating')
df.plot.pie(y='Flavor Rating', figsize=(13,7))
```
